## What helps a golfer win?
There are a handful of statistics in the dataset, all which have different implications about each golfer. Which of these stats, however, is most influential on winning? To explore this question, we need to:
- Examine the variation in each statistic, while keeping in mind what weight each one carries
- Study the implications of the top 3 placing for each golfer. Do better golfers place in the top 3 or top 10 more frequently? Or are they normally taking home 1st place.
- Explore outliers, are there top-level golfers who have been left off the higher rankings this year? Why would that be? On the other end, are there not so skilled players who have made their way to the top rankings?


```python
# variance examination

stds = pd.DataFrame({
        'Stat':data.columns[5:],
        'ST DEV':[data[col].std() for col in data.columns[5:]]
    })
stds = stds.sort_values(by='ST DEV')
Chart(stds[stds['ST DEV']<15]).mark_bar().encode(
        x='Stat',
        y='ST DEV')
```

I decided to rule out the standard deviation for Average Drive and Rounds (and others which are not even shown on the graph), since both are categories I would expect to have a much larger variation, with little importance in this study. 

Apart from those, Driving Percentage, Greens Percentage, and Scrambling Percentage had the highest variation, which is pretty interesting since they are all such different areas of a players game but all could be key to giving players an edge. Another aspect is strokes gained total having the highest variation, which will be analyzed later.

```python
# top 3 placing examination

data10 = data[data['Events']>0]
for place in ['1st','2nd','3rd']:
    plt.figure()
    plt.title(place)
    plt.scatter(data10[data10[place]>0][place], data10[data10[place]>0]['Events'])
```

Interesting observation here: of the top 3 placings, the 1st place winners have the least amount of top 10 finishes, meaning the more consistent golfers tend to land among 2nd or 3rd place, while the top spot goes to someone who performs less consistently.

```python
plt.xlabel('Average Score')
plt.ylabel('Top10 Placements')
plt.title('Score Less Important Than We Think?')
plt.scatter(data10['Average Score'], data10['Events'], s=data10['Rounds'])
```

The plot of average score against number of Top 10 placements is ALSO surprising. It shows a slight positive correlation, meaning players with higher scores place in the Top 10 more frequently? Strange (lower sores are btter in golf).

```python
for place in ['1st','2nd','3rd']:
    plt.figure()
    plt.title(place)
    plt.scatter(data10[data10[place]>0]['Average Score'], data10[data10[place]>0][place])
```

Based on these 3 visualizations, I've concluded that while 1st place winners score lower on average, 2nd and 3rd place winners have less of tendency for lower scores, and that players who place in the top 10's frequency of doing so has little to no tendency towards lower scores

```python
# Explore outliers
plt.title('Avg. Scores V. Appearances in Top 10')
plt.xlabel('Average Score')
plt.ylabel('Number of Appearances in Top 10')
plt.scatter(data['Average Score'], data['Events'], s=data['Rounds'])
plt.savefig('outliers1')
```

```python
data0 = data[data['Events']==0]
dataTop = data0.sort_values(by='Average Score', ascending=True).head(3)
```

```python
dataBot = data10.sort_values(by='Average Score', ascending=False).head(3)
```

```python
dataTop.head()
```

```python
plt.scatter(dataTop['Average Drive'], dataTop['Driving Percentage'])
plt.scatter(dataBot['Average Drive'], dataBot['Driving Percentage'])
plt.legend(['No Top 10 Placement', 'Top 10 Placement'])
plt.xlabel('Average Drive')
plt.ylabel('Driving Accuracy')
plt.title('Outliers')
plt.savefig('outliers2')
```

After comparing the golfers with the best 3 average scores with no top 10 placements against the golfers with the worst 3 average scores who have all placed in the top 10 (more than 10 time each, in fact), I found that the non-placing golfers surpssed the placing ones in almost every statistic. However, the golfers who did place in the top 10 had a higher average drive than those who didn't, as well as the highest average drive out of all 6 golfers. Additionally, one of the placing golfers had the second highest driving percentage out of all 6 golfers.

This is evidence that driving ability plays a major role in placing a top ranking, and should be weighted heavily.

## What about the course?
Obviously some golfers play better on different courses because they are all fit to play to different strengths (driving ability, putting consistency, scramble ability, etc). In order to make a more accurate model, I need to explore what strengths Erin Hill plays to and which ones it does not.

## Notes on Erin Hill

- Tough bunkers, hardly any flat bottoms
- Very wide fairways
- Lots of semi-blind shots
- Large, smooth, forgiving greens
- Longer than a typical US Open
- Scrambling ability may not provide as much of an edge

```python
course = pd.read_csv('course.csv').drop('Unnamed: 0', axis=1)
```

```python
course['Bunker'] = ['2' if 'bunkers' in desc else '1' if 'bunker' in desc else 0 for desc in course['Description']]
```

```python
Chart(course).mark_circle().encode(
    x='Hole',
    y='Distance',
    color='Par:N',
    opacity='Bunker')
```

This graph visualizes the distance for each hole of the course, categorized by Par level, and filled in relation to the intensity/frequency of bunkers. From it, I can conclude the Par 4s are longer than normal, so driving range will definitely be helpful. Additionally, most of the long Par 4s face less bunker intensity/frequency, giving more wiggle room to accuracy and less of an edge to scramble capability. However, the longer Par 5s will see more of an edge with scrambling capabilities, as their bunker intensity/frequency is higher than most holes.
