## Machine Learning
For my machine learning model, I would like to predict the winner of this year's (2017) US Open. Unfortunately, no US Open or any other major PGA event has been held at Erin Hill, so it is not possible to fit training data to account for elements from that course. Additionally, I believe fitting the data to winners of past US Opens would prove innacurate because they have all been played on vastly different courses which give all different golfers' styles an edge.

Instead, I will train a model to predict the number of times a golfer has placed in the Top 10 Rankings, and whether a top ranked golfer has ever won 1st place or not.

Then, I will use analysis to find the optimal player abilities for Erin Hill, and train a model to predict which golfer embodies these most.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from altair import *
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.cross_validation import cross_val_score
```

## Top 10 Model

```python
data_golfers = pd.read_csv('golfers.csv').drop('Total Adjustment', axis=1)
```

```python
# Linear Regression Model

data_golfers['Long Drive'] = [1 if avg > 290 else 0 for avg in data_golfers['Average Drive']]
model = LinearRegression()
X = data_golfers[['Average Score', 'Average Drive', 'Long Drive', 'Scrambling %', 'Rounds']]
model.fit(X, data_golfers['Events'])

print(-cross_val_score(model, X, data_golfers['Events'], cv=10, 
               scoring="neg_mean_squared_error").mean())
```

```python
y_new = model.predict(X)

plt.scatter(range(0, len(data_golfers)), data_golfers['Events'])
plt.scatter(range(0, len(data_golfers)), y_new)
plt.ylabel('Top 10 Appearances')
plt.legend(['Acutal', 'Predicted'])
plt.savefig('LinReg')
```

```python
# KNeighbors Model

error = []
for k in range(1,60):
    model = KNeighborsRegressor(k)
    X = data_golfers[['Average Score', 'Average Drive', 'Long Drive', 'Scrambling %', 'Rounds']]
    model.fit(X, data_golfers['Events'])
    error.append(-cross_val_score(model, X, data_golfers['Events'], cv=10, 
                scoring="neg_mean_squared_error").mean())
```

```python
plt.plot(range(1,60), error)
plt.xlabel('k')
plt.ylabel('Error')
```

```python
# Optimal k value is 33

k = error.index(min(error))+1
model = KNeighborsRegressor(k)
X = data_golfers[['Average Score', 'Average Drive', 'Long Drive', 'Scrambling %', 'Rounds']]
model.fit(X, data_golfers['Events'])
print(-cross_val_score(model, X, data_golfers['Events'], cv=10, 
            scoring="neg_mean_squared_error").mean())
```

The Linear Regression Model proved more accurate than the KNeighbors in predicting amount of Top 10 finishes for a professional golfer this year.

## 1st Place Winner?

```python
data_golfers['Won First'] = [1 if first > 0 else 0 for first in data_golfers['1st']]
```

```python
# Logistic Regression Model

dataTop = data_golfers[data_golfers['Events']>0]
model = LogisticRegression()
X = dataTop[['Events', 'Average Score', 'Average Drive', 'Long Drive', 'Scrambling %', 'Rounds']]
model.fit(X, dataTop['Won First'])

probs = model.predict_proba(X)[:,1]
plt.scatter(range(0, len(dataTop)), dataTop['Won First'])
plt.plot(range(0, len(dataTop)), probs, 'r-')
```

```python
print(cross_val_score(model, X, dataTop['Won First'], cv=10, 
               scoring="f1").mean())
```

```python
# KNeighbors Classifier

X = dataTop[['Events', 'Average Score', 'Average Drive', 'Long Drive', 'Scrambling %', 'Rounds']]

model = KNeighborsClassifier(30)
model.fit(X, dataTop['Won First'])

probs = model.predict_proba(X)[:,1]
plt.scatter(range(0, len(dataTop)), dataTop['Won First'])
plt.plot(range(0, len(dataTop)), probs, 'r-')
```

I found the Logistic Regression model to be more useful than the KNeighbors Classifier, although neither seems too accurate.

## US Open at Erin Hill
In order to predict the winner of the US open, I need to 1) find the optimal values of play ability in relation to the course and 2) create a model that will predict a golfer given these values

1) Finding optimal values

I found that at Erin Hill the fairways at Erin Hill are wide and forgiving, increasing the weight on driving distance and decreasing it on driving accuracy. I also found that the greens are smooth and forgiving, decreasing the weight of strokes-gained on puts. Somehow, I need to incorporate these factors into finding an optimal candidate.

```python
data_golfers['Top3'] = data_golfers['1st']+data_golfers['2nd']+data_golfers['3rd']
```

```python
dataTop = data_golfers[data_golfers['Events']>0]
```

To choose optimal values for the player, I'll take players with the top 5 amount of times having come in 1st, 2nd, or 3rd this year, and average their statistics

```python
# Optimal stats
# weight and unweight values as necessary by adding or subtracting 1 standard deviation

top10 = dataTop.sort_values(by='Top3', ascending=False).head(5)
score = top10['Average Score'].mean()
drive = top10['Average Drive'].mean()+9
drive_perc = top10['Driving Percentage'].mean()-5
SG_putting = top10['SG: Putting'].mean()-0.3
scramble = top10['Scrambling %'].mean()-3.8
```

```python
# Predict player with these stats
data_golfers = data_golfers[data_golfers['Name']!='Luke List'] # not participating in US Open
data_golfers = data_golfers[data_golfers['Name']!='Tony Finau'] # not participating in US Open


model = LogisticRegression() # Use logistic regressinon, as it resulted more accurate previously
X = data_golfers[['Average Score', 'Average Drive', 'Driving Percentage', 'SG: Putting', 'Scrambling %', 'Long Drive']]
model.fit(X, data_golfers['Name'])

model.predict([[score, drive, drive_perc, SG_putting, scramble, 1]]) # manually assign long drive
```

My predicted winner for this year's US Open is Brooks Koepka, we'll see what happens this weekend!
