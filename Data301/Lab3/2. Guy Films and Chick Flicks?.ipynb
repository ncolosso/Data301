## Part 1

The 1995 Alicia Silverstone movie _Clueless_ is regarded by some as the quintessential "chick flick". But do women really enjoy the movie more than men do? Conduct a hypothesis test __by simulation__ to test whether or not women rate the movie _significantly_ higher than men do.

Be sure to:
- state the null hypothesis clearly
- set up an appropriate simulation to obtain a $p$-value
- interpret the $p$-value in the context of the problem

_Optional_: To check your answer (and for extra practice), you may also want to carry out the hypothesis test by theory.

Our null hypothesis for this problem is that men and women enjoy the movie the same amount, meaning we would use 0 as our difference in ratings for men and women in our simulation.

```python
columns = "UserID::MovieID::Rating::Timestamp".split("::")
columns2 = "UserID::Gender::Age::Occupation::Zip-Code".split("::")
ratings = pd.read_table('/data/movielens/ratings.dat', sep='::', 
                        header=None, names=columns, engine='python')
users = pd.read_table('/data/movielens/users.dat', sep='::', 
                        header=None, names=columns2, engine='python')
```

```python
## MovieID for Clueless (1995) is 39, found using df in problem 1
data_merged = users.merge(ratings, how='left', on=["UserID"])
data_clueless = data_merged[data_merged["MovieID"] == 39]
data_clueless.head()
```

```python
model = BoxModel(box=list(data_clueless["Gender"]), size=len(data_clueless), replace=False)
model.sim(3)
```

```python
def calculate_difference(gender):
    gender = np.array(gender)
    return data_clueless[gender == "M"]["Rating"].mean() - data_clueless[gender == "F"]["Rating"].mean()
diffs = RV(model, calculate_difference)
sims = diffs.sim(10000)
```

```python
sims.plot(type="bar", bins=30)
```

```python
clueDiff = calculate_difference(data_clueless["Gender"])
clueDiff
```

```python
sims.count_leq(clueDiff)/10000
```

If we assume our null hypothesis that there truly is no difference in the average rating of "Clueless" between men and women, there is a 0% chance we would observe a difference in means of -0.312 from a given sample. Given this small p-value, we reject the null and conclude that on average Females rate this movie higher than Males do.

## Part 2 

Some people believe that James Bond films are targeted at men. Do men really enjoy Bond movies more than women? Let's examine the 1995 Bond movie _GoldenEye_. Conduct a hypothesis test **by theory** to test whether or not men rate _GoldenEye_ significantly higher than women do.

Be sure to:
- state the null hypothesis clearly
- do the appropriate calculations to obtain a $p$-value
- interpret the $p$-value in the context of the problem

__Optional:__ To check your answer (and for extra practice), you may also want to carry out the hypothesis test by simulation.

```python
## MovieID for GoldenEye (1995) is 10, found using df in problem 1
data_merged = users.merge(ratings, how='left', on=["UserID"])
data_golden = data_merged[data_merged["MovieID"] == 10]
data_golden.head()
```

```python
model = BoxModel(box=list(data_golden["Gender"]), size=len(data_golden), replace=False)
def calculate_difference2(gender):
    gender = np.array(gender)
    return data_golden[gender == "M"]["Rating"].mean() - data_golden[gender == "F"]["Rating"].mean()
diffs = RV(model, calculate_difference2)
sims = diffs.sim(10000)
sims.plot(type="bar", bins=30)
```

```python
goldDiff = calculate_difference2(data_golden["Gender"])
goldDiff
```

```python
sims.count_leq(goldDiff)/10000
```

If we assume our null hypothesis that there truly is no difference in the average rating of "GoldenEye" between men and women, there is a 85% chance we would observe a difference in means of 0.083 from a given sample. Given this large p-value, we can conclude that on average, there is no evidence of Men enjoying bond movies more than Women.
