```python
import pandas as pd

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
%matplotlib inline
```

```python
player_data = pd.read_csv("Players_2000s.csv")
```

```python
players = player_data[["player_id", "birth_country", "birth_state", "birth_city"]]
```

```python
city_loc = pd.read_csv("simplemaps-worldcities-basic.csv")
```

```python
merged_data = players.merge(city_loc, right_on=["city"], left_on=["birth_city"])
```

```python
m = Basemap(projection='robin',lon_0=0,resolution='c')
x, y = m(merged_data["lng"].values, merged_data["lat"].values)
plt.figure(figsize=(12,6))
m.drawlsmask()
m.drawcountries()
plt.plot(x, y, 'r.', markersize=2)
```
