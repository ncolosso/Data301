## Worked Example

I have a coin. I do not know if it is fair or not. I toss it 100 times and get 60 heads. Is the coin fair or not?

This is a question that we have answered before by simulation. Now let's answer this question by theory.

**Step 1.** First, let's set up the box, assuming the null hypothesis that the coin is fair. To use the Central Limit Theorem, we just need to know the mean and SD of the box.

```python
box = [0, 1]
mean(box), sd(box)
```

**Step 2.** By the Central Limit Theorem, the mean of 100 draws from this box is approximately $\textrm{Normal}(\mu, \frac{\sigma}{\sqrt{n}})$. Let's simulate many realizations of this mean and plot the distribution of means.

```python
model = Normal(mean=mean(box), sd=sd(box) / sqrt(100))
PercentHeads = RV(model)
sims = PercentHeads.sim(5000)
sims.plot()
```

**Step 3.** We want to know the probability that this mean is as large as $0.60$. Let's calculate this from our simulations. Compare this $p$-value with the one you got earlier, by simulating directly from the box.

```python
sims
```

```python
sims.count_geq(0.60) / 5000
```

**Step 4.** Alternatively, we could have converted $0.60$ to a $z$-score and compared it to a $\text{Normal}(0, 1)$ distribution.

```python
model = Normal(mean=0, sd=1)
Z = RV(model)
sims = Z.sim(5000)

z = (0.60 - mean(box)) / (sd(box) / sqrt(100))
print(z)
sims.count_geq(z) / 5000
```

## Problem 1

We have an instrument that measures the height of a mountain. The measurements 
are centered around the true height, but with a standard deviation of 10 feet.

We make 15 measurements of the height of Mt. Everest, 
and find that the average is 29,023 feet. The official height is 29,029 feet. Has 
Mt. Everest gotten shorter?

```python
evmodel = RV(Normal(mean=29029, sd=10/sqrt(15)))
evsims = evmodel.sim(1000)
evsims.plot()
```

```python
evsims.count_leq(29023)/1000
```

Assuming the true height of everest has not changed and is still 29,029 ft, there is a 1.3% chance that we would observe an average measurment of less than or equal to 29,023 from a sample. Based on this, we reject the null hypothesis and conclude that the height of mount everest has changed. If the height hasn't changed, we have to accept that a 1 in 100 event has just happened.

```python
Z = RV(Normal(mean=0, sd=1))
sims = Z.sim(1000)
```

```python
zSamp = (29023 - 29029)/(10/sqrt(15))
zSamp
```

```python
sims.count_leq(zSamp)/1000
```
