# Clustering in Scikit-Learn

Now we will look at how to cluster data using scikit-learn. The clustering models can all be found in the `sklearn.cluster` module. Clustering follows the same general scikit-learn API, with a few differences:

1. Specify the model.
2. Fit the model with `model.fit(X)`. (Note that there's no `y`.)
3. Assign new data to clusters with `model.predict(X_new)`.

_Note:_ To get the clusters for the training data, you can combine steps 2 and 3 using `model.fit_predict(X)`.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

## K-Means Clustering of Wines Data

Cluster the wines data into two clusters using $k$-means clustering. Which wines get assigned to cluster 0? And which get assigned to cluster 1?

```python
wines_data = pd.read_csv("/data/wines.csv", sep=";")
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=2)
X = (wines_data - wines_data.mean())/ wines_data.std()
#model.fit(X)

model.fit_predict(X)
```

## K-Means Clustering of Handwritten Digits

Now let's cluster the handwritten digits data set.

```python
digits_train = pd.read_csv("/data/digits_train.csv")
```

Each row of this data frame is a length-784 vector that represents a 28 x 28 image. You can view one of these images by reshaping the vector into a 28 x 28 matrix and using `plt.imshow()`.

```python
plt.imshow(digits_train.loc[0].values.reshape((28, 28)), cmap="Greys")
```

Now let's cluster the handwritten digits into 10 clusters. Note that the cluster centers (`model.cluster_centers_`) are each length-784 vectors. Reshape them into 28 x 28 images and visualize them.

```python
# YOUR CODE HERE
model = KMeans(n_clusters=10)
model.fit_predict(digits_train)
```
