# Basemap and Dot Maps

A [dot map](http://weeklymap.org/basics.html) depicts the geographic locations of the observations in our data. Let's make a dot map showing the location of the earthquakes in the United States in the past 24 hours. We will use a package called Basemap that is part of the Matplotlib toolkit.

```python
import pandas as pd

import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
%matplotlib inline
```

```python
data = pd.read_csv("/data/earthquakes.csv")
data.head()
```

First, we set up the map, which consists of a specification of the area to plot and a [projection](http://matplotlib.org/basemap/users/mapsetup.html). There are two main ways to specify what area to plot:

- specify the central latitude and longitude (`lat_0` and `lon_0`) and the `width` and `height` (in kilometers)
- specify the latitude and longitude of the lower left corner (`llcrnrlon` and `llcrnrlat`) and the upper right corner (`urcrnrlon` and `urcrnrlat`).

A common projection that is used for maps of the United States is the [Lambert conformal conic](https://www.wired.com/2013/11/projection-lambert-conformal-conic/). This projection keeps two "standard" parallels undistorted. For the U.S., it is typical to keep the 33${}^\circ$ and 45${}^\circ$ parallels (`lat_1` and `lat_2`) undistorted. We also have to specify a central meridian (`lon_0`).

```python
m = Basemap(llcrnrlon=-119, llcrnrlat=22, urcrnrlon=-64, urcrnrlat=49,
            projection="lcc", lat_1=33, lat_2=45, lon_0=-95)
```

Now let's draw a map background. There are a number of options:

- `.drawlsmask()`: land-sea mask which draws gray for land, white for water
- `.bluemarble()`: NASA Blue Marble image
- `.shadedrelief()`: relief that shows topographical features
- `.etopo()`: ETOPO relief that shows topographical features

```python
m.bluemarble()
```

You might want to add country or state boundaries. There are a number of methods beginning with `.draw*()`, including:

- `.drawcountries()`
- `.drawstates()`
- `.drawcoastlines()`
- `.drawcounties()`

All of these functions take optional arguments, like color. To find similar functions and learn how they work, use tab completion and read the documentation.

```python
m.shadedrelief()
m.drawstates()
m.drawcountries()
m.drawcoastlines()
```

Now let's add points to the plot. One way to do this is to simply call `plt.plot()`, but we have to remember to convert the longitude and latitude to the x- and y- coordinates of the map. The `Basemap` object that we have can do this conversion for us.

```python
# .values converts the Pandas series to a Numpy array
x, y = m(data["longitude"].values, data["latitude"].values)

# make the map with the points overlaid
m.shadedrelief()
plt.plot(x, y, 'k.')
```

Alternatively, we can use the `.plot()` method of the `Basemap` object, which mirrors Matplotlib's `.plot()` function, but with an additional argument `latlon`, that specifies whether `x` and `y` are specified as longitude and latitude.

```python
m.shadedrelief()
m.plot(data["longitude"].values, data["latitude"].values, 'k.',
       latlon=True)
```

We might want to use the size of each point to indicate the magnitude of each eartquake. Use the `.scatter()` method of the `Basemap` object to set of the size each point based on the magnitude (`mag`).

```python
data = data[data["mag"]>=0]
```

```python
m.shadedrelief()
m.scatter(x=data["longitude"].values, 
          y=data["latitude"].values, 
          s=data["mag"].values,
          latlon=True,
          c='black')
```
