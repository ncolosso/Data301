# Polynomial Regression

```python
import pandas as pd
%matplotlib inline

data = pd.read_csv("/data/harris.csv")
data.plot.scatter(x="Exper", y="Bsal")
```

Looking at the scatterplot above, it seems clear that the relationship between experience and starting salary is not linear. Suppose we wanted to fit a polynomial curve to the data:

$$ f(\tt{Exper}) = \beta_0 + \beta_1 \tt{Exper} + \beta_2 \tt{Exper}^2. $$

How could we fit such a model using linear regression? Fit the model, and show the fitted curve overlaid on a scatterplot like the one above.

```python
# YOUR CODE HERE
data['Exper2'] = data['Exper']**2

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(data[['Exper', 'Exper2']], data['Bsal'])
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_new = np.arange(0, 400)
X_new = pd.DataFrame({
        'Exper': x_new,
        'Exper2': x_new**2,
    })
y_new = model.predict(X_new)

data.plot.scatter(x='Exper', y='Bsal')
plt.plot(x_new, y_new, 'r-')
```
